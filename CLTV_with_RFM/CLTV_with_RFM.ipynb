```python
df = df[~df["Invoice"].astype(str).str.contains("C")]
```

```python
df["Total_Price"] = df["Quantity"] * df["Price"]
```

```python
import datetime as dt
```

```python
today_date = dt.datetime(2011, 12, 11)
```

```python

rfm = df.groupby("Customer ID").agg({"InvoiceDate":lambda x: (today_date-x.max()).days,"Invoice":lambda x: x.nunique(),"Total_Price":lambda x: x.sum()})
```

```python
rfm.columns = ["recency","frequency","monetary"]
```

```python
rfm
```

```python
rfm["recency_score"] = pd.qcut(rfm["recency"], 5, labels= [5,4,3,2,1])
rfm["frequency_score"] = pd.qcut(rfm["frequency"].rank(method = "first"), 5, labels= [1,2,3,4,5])
rfm["monetary_score"] = pd.qcut(rfm["monetary"], 5, labels= [1,2,3,4,5])
```

```python
rfm
```

```python
rfm["RF_SCORE"] = (rfm["recency_score"].astype(str)+ rfm["frequency_score"].astype(str))
```

```python
rfm
```

```python
seg_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2][3-4]': 'at_Risk',
    r'[1-2]5': 'cant_loose',
    r'3[1-2]': 'about_to_sleep',
    r'33': 'need_attention',
    r'[3-4][4-5]': 'loyal_customers',
    r'41': 'promising',
    r'51': 'new_customers',
    r'[4-5][2-3]': 'potential_loyalists',
    r'5[4-5]': 'champions'}
```

```python
rfm['segment'] = rfm['RF_SCORE'].replace(seg_map, regex=True)
```

```python
rfm
```

```python
rfm.groupby("segment").agg(["min","mean","max","count"])
```

```python
loyal_customers_ID = rfm[rfm["segment"] == "loyal_customers"]
```

```python
loyal_customers_ID = loyal_customers_ID.reset_index()
```

```python
loyal_customers_ID["Customer ID"].to_csv("loyal_customers_ID.csv")
```
